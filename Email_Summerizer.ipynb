```python
import os
import json
from langchain_community.document_loaders import PyPDFLoader
from langchain_groq.chat_models import ChatGroq
from langchain.prompts import PromptTemplate
```

```python
groq_api_key = os.getenv("GROQ_API_KEY")

llm = ChatGroq(
    model_name="llama3-8b-8192",  # Best balance for summarization
    groq_api_key=groq_api_key,
    temperature=0.2
)
```

```python
loader = PyPDFLoader("C:/Users/Admin/Desktop/AI-ML_QB/Email_Tasks/AI_Email_Dataset.pdf")  # Path to your PDF
pages = loader.load()
```

```python
subject=str(input("Enter Subject"))
to = str(input("Enter Recipent"))
From = str(input("Enter you email"))
body = str(input("enter you body"))
```

```python
## take data from user
email_text = f"""
Subject:{subject}
To:{to}
From:{From}

body:{body}
"""
```

```python
prompt = PromptTemplate(
    input_variables=["email_text"],
    template="""
You are an expert at summarizing emails.

Given the full raw email below, extract the following in JSON format:
- "from": sender's email address
- "to": recipient's email address
- "subject_summary": a clean, brief summary of the subject
- "body_summary": a concise summary of the body

Email:
{email_text}

Return ONLY valid JSON in this format:
{{
  "from": "...",
  "to": "...",
  "subject_summary": "...",
  "body_summary": "..."
}}
"""
)
```

```python
summaries = []

for i, page in enumerate(pages):
    email_text = page.page_content.strip()

    formatted_prompt = prompt.format(email_text=email_text)
    response = llm.invoke(formatted_prompt)
    print(response.content)
```

```python
with open("email_summaries.json", "w") as f:
    json.dump(summaries, f)
```
